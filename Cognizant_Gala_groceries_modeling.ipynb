Converting datatype of timestamp column using function

```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):

  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

#Merging the data
In order to include all the data within a predictive model, we need to merge all the 3 datasets into 1 dataframe.

The client indicates that they want the model to predict on an hourly basis.looking at the 3 datasets,the only column we can use to merge the 3 datasets together is 'timestamp'. 
So first transform the timestamp column in all 3 datasets to be based on the hour of the day, then we can merge the datasets together.

```python
sales_df.head()
```

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
```

```python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
```

```python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
```

Now aggregating the datasets to combine rows which have the same value for timestamp

```python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
```

we now have an aggregated sales data where each row represents a unique combination of hour during which the sales took place from that weeks worth of data and product_id.

```python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
```

This shows the average stock percentage of each product at unique hours within the week of sample data

```python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

For the temperature data, product_id column does not exist in this table, so we need to group by timestamp and aggregate the temperature. This gives us the average temperature of the storage facility where the product is stored in the warehouse by unique hours during the week.

```python
merged_df = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.merge(temp_agg, on='timestamp', how='left')
merged_df.head()
```

```python
merged_df.info()
```

We can see we have some null values in the quantity column. we can assume that there if there is null values it represents that there are 0 sales of this product within this hour.Lets fill this column null values with 0.

```python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
```

Combining some more features on the table, including category and unit price.

```python
product_categories = sales_df[['product_id', 'category']]
product_categories = product_categories.drop_duplicates()

product_price = sales_df[['product_id', 'unit_price']]
product_price = product_price.drop_duplicates()
```

```python
merged_df = merged_df.merge(product_categories, on="product_id", how="left")
merged_df.head()
```

```python
merged_df = merged_df.merge(product_price, on="product_id", how="left")
merged_df.head()
```

```python
merged_df.info()
```

#Feature Engineering

The timestamp column in its current form is not very useful for a ml model.first, exploding the timestamp column into day of week, day of month and hour

```python
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
```

Next the category column, Its in a categorical form. we can convert it to numerical form by creating dummy variables.

```python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

the product_id is simply an ID column, it will add no value by including it in the predictive model. Hence, we shall remove it from the modeling process.

```python
merged_df.drop(columns=['product_id'], inplace=True)
merged_df.head()
```

#Modelling

We will use a supervised machine learning model, and we will use `estimated_stock_pct` as the target variable, since the problem statement was focused on being able to predict the stock levels of products on an hourly basis.

Creating target variable 'y' and independent variable 'x'

```python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

we will use a `K-fold` strategy to train the machine learning model on `K` random samples of the data.

```python
K = 10
split = 0.75
```

We are going to use RandomForestRegressor Model.We are using a `regression` algorithm here because we are predicting a continuous numeric variable, that is, `estimated_stock_pct`.

```python
!pip install scikit-learn
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
accuracy = []

for fold in range(0, K):

  # Instantiate algorithm
  model = RandomForestRegressor()
  scaler = StandardScaler()

  # Create training and test samples
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)

  # Scale X data, we scale the data because it helps the algorithm to converge
  # and helps the algorithm to not be greedy with large values
  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)

  # Train model
  trained_model = model.fit(X_train, y_train)

  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)

  # Compute accuracy, using mean absolute error
  mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
  accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
```
